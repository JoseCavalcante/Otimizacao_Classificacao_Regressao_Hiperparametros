**solver (solucionador) {'newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'}, default='lbfgs'**  **:  técnica matemática**

Conjuntos de dados pequenos: 'liblinear'

Conjuntos de dados grandes: 'sag' e 'saga'

Apenas 'newton-cg', 'sag', 'saga' e 'lbfgs' para multinomial;

A escolha do algoritmo depende da penalidade escolhida:

'newton-cg' - ['l2', 'nenhum']

'lbfgs' - ['l2', 'nenhum']

'liblinear' - ['l1', 'l2']

'sag' - ['l2', 'nenhum']

'saga' - ['elasticnet', 'l1', 'l2', 'nenhum']

Explicações mais detalhadas dos solucionadores: https://stackoverflow.com/questions/38640109/logistic-regression-python-solvers-definitions

**tol (Tolerância para critérios de parada):** float, default=1e-4


**max_iter (Número máximo de iterações para os solucionadores convergirem): int, default=100**


**C (PARÂMETRO C) float, default = 1,0**

Deve ser positivo. Quanto menor o valor mais forte a regularização .
